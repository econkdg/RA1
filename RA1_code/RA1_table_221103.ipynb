## **1. data_221020**

```python
# import the data

data_221020 = pd.read_excel('data_221020.xlsx').drop(columns=['Unnamed: 0'])
data_221020['quarter'] = pd.to_datetime(data_221020['quarter'])
data_221020['quarter'] = data_221020['quarter'].dt.to_period('Q')
```

```python
# quarterly dataset 221020

data_221020 = data_221020.sort_values(by=['isocode', 'quarter']).reset_index(drop=True)

data_221020
```

## **2. tab_4_vars_221103**

```python
# 4 variables: (1) ggcon, (2) gdp, (3) reer, (4) ca

tab_4_vars_221103 = pd.DataFrame(data_221020['isocode'].unique().reshape(-1, 1)).rename(columns={0:'isocode'})

tab_4_vars_221103['begin'] = np.nan
tab_4_vars_221103['end'] = np.nan

for i in range(0, data_221020['isocode'].unique().shape[0]):
  
    # generate the count matrix for quarterly data by isocode
    globals()['cond_isocode_{}'.format(i+1)] = (data_221020['isocode'] == data_221020['isocode'].unique()[i])
    globals()['df_isocode_{}'.format(i+1)] = data_221020.loc[globals()['cond_isocode_{}'.format(i+1)]]
    globals()['df_isocode_count_T_{}'.format(i+1)] = globals()['df_isocode_{}'.format(i+1)].groupby(globals()['df_isocode_{}'.format(i+1)]['quarter']).count()

    # condition for 4 variables
    globals()['cond_ggcon_{}'.format(i+1)] = (globals()['df_isocode_count_T_{}'.format(i+1)]['ggcon'] == 0)
    globals()['cond_gdp_{}'.format(i+1)] = (globals()['df_isocode_count_T_{}'.format(i+1)]['gdp'] == 0)
    globals()['cond_reer_{}'.format(i+1)] = (globals()['df_isocode_count_T_{}'.format(i+1)]['reer'] == 0)
    globals()['cond_ca_{}'.format(i+1)] = (globals()['df_isocode_count_T_{}'.format(i+1)]['ca'] == 0)

    # filter for 4 variables
    globals()['df_isocode_count_T_{}'.format(i+1)] = globals()['df_isocode_count_T_{}'.format(i+1)].loc[~globals()['cond_ggcon_{}'.format(i+1)] & ~globals()['cond_gdp_{}'.format(i+1)] & ~globals()['cond_reer_{}'.format(i+1)] & ~globals()['cond_ca_{}'.format(i+1)]]
    globals()['df_isocode_count_T_{}'.format(i+1)] = globals()['df_isocode_count_T_{}'.format(i+1)].reset_index()

    # print the results
    print('country', data_221020['country'].unique()[i])
    print('isocode', data_221020['isocode'].unique()[i])
    print('start:', globals()['df_isocode_count_T_{}'.format(i+1)]['quarter'].unique().min())
    print('end:', globals()['df_isocode_count_T_{}'.format(i+1)]['quarter'].unique().max())

    # replace the value of both begin and end
    globals()['cond_begin_end_{}'.format(i+1)] = (tab_4_vars_221103['isocode'] == data_221020['isocode'].unique()[i])

    tab_4_vars_221103.loc[globals()['cond_begin_end_{}'.format(i+1)], 'begin'] = globals()['df_isocode_count_T_{}'.format(i+1)]['quarter'].unique().min()
    tab_4_vars_221103.loc[globals()['cond_begin_end_{}'.format(i+1)], 'end'] = globals()['df_isocode_count_T_{}'.format(i+1)]['quarter'].unique().max()

    print('')
```

```python
# table (4 variables)

tab_4_vars_221103 = tab_4_vars_221103.transpose()
tab_4_vars_221103
```

```python
# export the table (4 variables)

tab_4_vars_221103.to_excel(excel_writer='tab_4_vars_221103.xlsx')
```
