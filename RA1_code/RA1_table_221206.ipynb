## **1. data preprocessing**

```python
# import the data

df_m_infl = pd.read_excel('df_m_infl.xlsx')
```

```python
# 1st dataframe

df_m_infl_1 = df_m_infl.iloc[0:3].set_index([0]).transpose().reset_index()
df_m_infl_1
```

```python
# 2nd dataframe

df_m_infl_2 = df_m_infl.iloc[3:].set_index([0]).transpose().reset_index()
df_m_infl_2
```

```python
# generate the long format table

df_m_infl = pd.merge(df_m_infl_1, df_m_infl_2, how='outer', on=['index']).drop(columns=['index'])
df_m_infl = df_m_infl.set_index(['country', 'sources', 'variable'])
df_m_infl = df_m_infl.transpose().reset_index().rename(columns={0:'month'}).set_index(['month'])
df_m_infl = pd.DataFrame(df_m_infl.stack(level='country').stack(level='sources').stack(level='variable'))
df_m_infl = df_m_infl.rename(columns={0:'value'})
df_m_infl = df_m_infl.reset_index()
df_m_infl = df_m_infl[['sources', 'variable', 'country', 'month', 'value']]
df_m_infl = df_m_infl.sort_values(by=['sources', 'variable', 'country', 'month']).reset_index(drop=True)

df_m_infl['month'] = pd.to_datetime(df_m_infl['month'])
df_m_infl['month'] = df_m_infl['month'].dt.to_period('M')

df_m_infl
```

```python
# generate the sources:variable

# concat_columns func.
def concat_columns(col_1, col_2):

    if not pd.isna(col_2):
       
       result = str(col_1) + ':' + str(col_2)

    return result

df_m_infl['sources:variable'] = df_m_infl.apply(lambda x: concat_columns(x['sources'], x['variable']), axis=1)

df_m_infl = df_m_infl[['sources', 'variable', 'sources:variable', 'country', 'month', 'value']]
df_m_infl
```

## **2. python code**

```python
# generate the sample period table

sources_variable_list = list(df_m_infl['sources:variable'].unique())

sources_variable_num = 1

for sources_variable in sources_variable_list:

    cond = (df_m_infl['sources:variable'] == sources_variable)

    data = df_m_infl.loc[cond].reset_index(drop=True)

    for i in range(0, data['country'].unique().shape[0]):
  
        # generate the count matrix for monthly data by country
        globals()['cond_country_{}'.format(i+1)] = (data['country'] == data['country'].unique()[i])
        globals()['df_country_{}'.format(i+1)] = data.loc[globals()['cond_country_{}'.format(i+1)]]
        globals()['df_country_count_T_{}'.format(i+1)] = globals()['df_country_{}'.format(i+1)].groupby(globals()['df_country_{}'.format(i+1)]['month']).count()

        # condition for value
        globals()['cond_value_{}'.format(i+1)] = (globals()['df_country_count_T_{}'.format(i+1)]['value'] == 0)

        # filter for value
        globals()['df_country_count_T_{}'.format(i+1)] = globals()['df_country_count_T_{}'.format(i+1)].loc[~globals()['cond_value_{}'.format(i+1)]]
        globals()['df_country_count_T_{}'.format(i+1)] = globals()['df_country_count_T_{}'.format(i+1)].reset_index()

        globals()['tab_{}'.format(i+1)] = pd.DataFrame(np.array([data['country'].unique()[i], 
                                                                 str(globals()['df_country_count_T_{}'.format(i+1)]['month'].unique().min()) + ':' + str(globals()['df_country_count_T_{}'.format(i+1)]['month'].unique().max()),
                                                                 globals()['df_country_count_T_{}'.format(i+1)]['month'].count()]).reshape(1, -1)
                                                                 ).rename(columns={0: 'country', 
                                                                                   1: 'period', 
                                                                                   2: 'obs'}).reset_index(drop=True)

    # print(sources_variable)

    if data['country'].unique().shape[0] == 1:

       globals()['tab'] = globals()['tab_{}'.format(1)]

       globals()['table_{}'.format(sources_variable_num)] = pd.DataFrame(globals()['tab']).reset_index(drop=True)
    
    elif data['country'].unique().shape[0] == 2:

       globals()['tab'] = pd.concat([globals()['tab_{}'.format(1)], globals()['tab_{}'.format(2)]])

       globals()['table_{}'.format(sources_variable_num)] = pd.DataFrame(globals()['tab']).reset_index(drop=True)

    elif data['country'].unique().shape[0] >= 3:

       globals()['tab'] = pd.concat([globals()['tab_{}'.format(1)], globals()['tab_{}'.format(2)]])

       for j in range(3, data['country'].unique().shape[0]+1):
           globals()['tab'] = pd.concat([globals()['tab'], globals()['tab_{}'.format(j)]])

       globals()['table_{}'.format(sources_variable_num)] = pd.DataFrame(globals()['tab']).reset_index(drop=True)
    
    sources_variable_num += 1
```
