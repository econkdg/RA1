## **2. python code**

```python
# 4 variables: (1) ggcon, (2) gdp, (3) reer, (4) ca

# import data_221115 as data
data = data_221115

def df():

    for i in range(0, data['isocode'].unique().shape[0]):
  
        # generate the count matrix for quarterly data by isocode
        globals()['cond_isocode_{}'.format(i+1)] = (data['isocode'] == data['isocode'].unique()[i])
        globals()['df_isocode_{}'.format(i+1)] = data.loc[globals()['cond_isocode_{}'.format(i+1)]]
        globals()['df_isocode_count_T_{}'.format(i+1)] = globals()['df_isocode_{}'.format(i+1)].groupby(globals()['df_isocode_{}'.format(i+1)]['quarter']).count()

        # condition for 4 variables
        globals()['cond_ggcon_{}'.format(i+1)] = (globals()['df_isocode_count_T_{}'.format(i+1)]['ggcon'] == 0)
        globals()['cond_gdp_{}'.format(i+1)] = (globals()['df_isocode_count_T_{}'.format(i+1)]['gdp'] == 0)
        globals()['cond_reer_{}'.format(i+1)] = (globals()['df_isocode_count_T_{}'.format(i+1)]['reer'] == 0)
        globals()['cond_ca_{}'.format(i+1)] = (globals()['df_isocode_count_T_{}'.format(i+1)]['ca'] == 0)

        # filter for 4 variables
        globals()['df_isocode_count_T_{}'.format(i+1)] = globals()['df_isocode_count_T_{}'.format(i+1)].loc[~globals()['cond_ggcon_{}'.format(i+1)] & ~globals()['cond_gdp_{}'.format(i+1)] & ~globals()['cond_reer_{}'.format(i+1)] & ~globals()['cond_ca_{}'.format(i+1)]]
        globals()['df_isocode_count_T_{}'.format(i+1)] = globals()['df_isocode_count_T_{}'.format(i+1)].reset_index()

        globals()['cond_T_min_{}'.format(i+1)] = (data['quarter'] >= globals()['df_isocode_count_T_{}'.format(i+1)]['quarter'].unique().min())
        globals()['cond_T_max_{}'.format(i+1)] = (data['quarter'] <= globals()['df_isocode_count_T_{}'.format(i+1)]['quarter'].unique().max())
        globals()['df_isocode_{}'.format(i+1)] = globals()['df_isocode_{}'.format(i+1)].loc[(globals()['cond_T_min_{}'.format(i+1)] & globals()['cond_T_max_{}'.format(i+1)])].reset_index(drop=True)

    globals()['df_4_vars'] = pd.concat([globals()['df_isocode_{}'.format(1)], globals()['df_isocode_{}'.format(2)]])

    for j in range(3, data['isocode'].unique().shape[0]+1):
        
        globals()['df_4_vars'] = pd.concat([globals()['df_4_vars'], globals()['df_isocode_{}'.format(j)]])

    df_4_vars = pd.DataFrame(globals()['df_4_vars']).reset_index(drop=True)

    # cond_isocode = (df_4_vars['isocode'] == isocode)
    # df_4_vars = df_4_vars.loc[cond_isocode]

    return df_4_vars
```

```python
# period func.
def period(data, var):

    data = data

    for i in range(0, data['isocode'].unique().shape[0]):
  
        # generate the count matrix for quarterly data by isocode
        globals()['cond_isocode_{}'.format(i+1)] = (data['isocode'] == data['isocode'].unique()[i])
        globals()['df_isocode_{}'.format(i+1)] = data.loc[globals()['cond_isocode_{}'.format(i+1)]]
        globals()['df_isocode_count_T_{}'.format(i+1)] = globals()['df_isocode_{}'.format(i+1)].groupby(globals()['df_isocode_{}'.format(i+1)]['quarter']).count()

        # condition for var.
        globals()['cond_{}'.format(var) + '_{}'.format(i+1)] = (globals()['df_isocode_count_T_{}'.format(i+1)][var] == 0)

        # filter for var.
        globals()['df_isocode_count_T_{}'.format(i+1)] = globals()['df_isocode_count_T_{}'.format(i+1)].loc[~globals()['cond_{}'.format(var) + '_{}'.format(i+1)]]
        globals()['df_isocode_count_T_{}'.format(i+1)] = globals()['df_isocode_count_T_{}'.format(i+1)].reset_index()

        globals()['tab_4_vars_{}'.format(i+1)] = pd.DataFrame(np.array([data['country'].unique()[i], 
                                                                        data['isocode'].unique()[i], 
                                                                        str(globals()['df_isocode_count_T_{}'.format(i+1)]['quarter'].unique().min()) + '-' + str(globals()['df_isocode_count_T_{}'.format(i+1)]['quarter'].unique().max()),
                                                                        globals()['df_isocode_count_T_{}'.format(i+1)]['quarter'].count()]).reshape(1, -1)
                                                                        ).rename(columns={0: 'country', 
                                                                                          1: 'isocode', 
                                                                                          2: 'period', 
                                                                                          3: 'obs'}).reset_index(drop=True)

    globals()['tab_4_vars'] = pd.concat([globals()['tab_4_vars_{}'.format(1)], globals()['tab_4_vars_{}'.format(2)]])

    for j in range(3, data['isocode'].unique().shape[0]+1):
        
        globals()['tab_4_vars'] = pd.concat([globals()['tab_4_vars'], globals()['tab_4_vars_{}'.format(j)]])

    tab_4_vars = pd.DataFrame(globals()['tab_4_vars']).reset_index(drop=True)

    return tab_4_vars

# advanced
def tab_advanced(data, var):

    advanced_list = ['AUS', 'AUT', 'BEL', 'CAN', 'CHE', 'CZE', 'DEU', 'DNK', 'ESP', 'EST', 
                     'FIN', 'FRA', 'GBR', 'GRC', 'HKG', 'IRL', 'ISR', 'ITA', 'JPN', 'KOR', 
                     'LTU', 'LVA', 'NLD', 'NOR', 'NZL', 'PRT', 'SVK', 'SVN', 'SWE', 'USA']

    for i in range(0, len(advanced_list)):
  
        globals()['cond_N{}'.format(i+1)] = (period(data, var)['isocode'] == advanced_list[i])

    tab_4_vars_advanced = period(data, var).loc[cond_N1 | cond_N2 | cond_N3 | cond_N4 | cond_N5 | cond_N6 | cond_N7 | cond_N8 | cond_N9 | cond_N10 |
                                                cond_N11 | cond_N12 | cond_N13 | cond_N14 | cond_N15 | cond_N16 | cond_N17 | cond_N18 | cond_N19 | cond_N20 |
                                                cond_N21 | cond_N22 | cond_N23 | cond_N24 | cond_N25 | cond_N26 | cond_N27 | cond_N28 | cond_N29 | cond_N30]
 
    tab_4_vars_advanced = tab_4_vars_advanced.sort_values(by=['isocode']).reset_index(drop=True)
 
    return tab_4_vars_advanced

# developing
def tab_developing(data, var):

    developing_list = ['ARG', 'BGR', 'BRA', 'CHL', 'COL', 'HRV', 'HUN', 'IDN', 'IND', 'MEX', 
                       'POL', 'ROU', 'RUS', 'THA', 'TUR', 'ZAF']

    for i in range(0, len(developing_list)):
  
        globals()['cond_N{}'.format(i+1)] = (period(data, var)['isocode'] == developing_list[i])

    tab_4_vars_developing = period(data, var).loc[cond_N1 | cond_N2 | cond_N3 | cond_N4 | cond_N5 | cond_N6 | cond_N7 | cond_N8 | cond_N9 | cond_N10 |
                                                  cond_N11 | cond_N12 | cond_N13 | cond_N14 | cond_N15 | cond_N16]

    tab_4_vars_developing = tab_4_vars_developing.sort_values(by=['isocode']).reset_index(drop=True)

    return tab_4_vars_developing
```
